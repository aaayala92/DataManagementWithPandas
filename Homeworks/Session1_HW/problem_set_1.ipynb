```python
# write your code here 

import pandas as pd

#1. Read the data contained in file SKU.xlsx in a DataFrame.

sku = pd.read_excel("SKU.xlsx")
```

```python
#2. Display the first 5 rows of and the last 3 rows of the dataset. 
sku.head()
```

```python
sku.tail(3)
```

```python
#3. How many columns does the DataFrame contain? 
tot_row, tot_column = sku.shape

print('SKU dataset has', tot_column, 'columns')
```

```python
#4. Create and add a new column called “pallet_value” (pallet_value =  unitprice x units_per_pallet).

sku['pallet_value'] = sku['unitprice']*sku['units_per_pallet']

#checking output
sku.head(3)
```

```python
#5. Create a new DataFrame with only SKUs whose pallet_value is higher than 200.

sku_200palletvalue = sku[sku['pallet_value']>200]

#checking output
sku_200palletvalue.sort_values(by=["pallet_value"])
```

```python
#6. Sort the DataFrame by descending order of pallet_value.
sku_200palletvalue.sort_values(by=['pallet_value'], ascending = False, inplace = True)

#checking output
sku_200palletvalue
```

```python
#7. Write the resulting DataFrame to an Excel file. Do not include the index.
sku_200palletvalue.to_excel('sku_200palletvalue.xlsx', index = False)
```
