```python
# write your code for in-class activity 3 here 


co2_2017 = pd.read_excel('CO2_emissions_2017.xlsx')

co2_2017.info()
```

```python
# 1. Create a column with the average C02 emissions
co2_2017['Avg_CO2']=co2_2017['CO2_emissions'].mean()
co2_2017
```

```python
# 2. Create a column with the standard deviation of CO2 emissions

co2_2017['Std_CO2']=co2_2017['CO2_emissions'].std()
co2_2017
```

```python
# 3. Create a column calculating the standardized C02 emissions: (value-average)/standard deviation
co2_2017['Standarized_CO2']=(co2_2017['CO2_emissions']-co2_2017['Avg_CO2'])/co2_2017['Std_CO2']
co2_2017
```

```python
# 4. Identify the vehicle (ID) with the largest standardized C02 emissions. What is the maximum value?
co2_2017['Standarized_CO2'].max()
```

```python
co2_2017[co2_2017['Standarized_CO2']==co2_2017['Standarized_CO2'].max()]
```

```python
co2_2017.loc[co2_2017['Standarized_CO2'].idxmax()]
```

```python
co2_2017.sort_values(by=['Standarized_CO2'],ascending = False)
```
