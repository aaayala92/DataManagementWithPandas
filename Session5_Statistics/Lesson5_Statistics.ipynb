# Statistical Analysis

### Session 1 - Statistical Analysis

Exploratory Data Analysis

### Machine Learning vs Statistics

- Machine Learning: designed to make accurate predictions. Called blackbox in that sense

- Statistical Models: designed for inference about relationship variabels. Relationship between variables

### Statistical Thinking Procedure

1. Describe the problem (Does variable A affect variable B)
2. Collect Data
3. Visualize the data
4. Look for outliers, EDA

### Variability

- Successive observation that do not produce exactly same results
- We need a variability benchmark to decide if it makes a difference or not

### Descriptive Statistics

- With a Sample we try to understand the population
- The sample must representative

### Data Types

- Categorical :['Nominal','Ordinal','Binary']
- Numerical: ['Continuous','Discrete']

- Poisson Distribution is to predict a count, discrete.

### How to use Sample to learn from population

- with 30 samples, that are representative, we can estimate the mean. True mean is miu, sample mean is Xbar. true variance is sigma, sample variance is s.

### Data Summaries

- range is also affected by outliers
- median is not affected by outliers. it is 50% of the data
- if mean and median are the same you have a clean normal distribution
- InterQuartile range is not heavily inference by outliers

### EDA

- Visualize the data
- Correlation Matrix
- Measure statistics such as mean, median, std
- Dimensionality reduction
- K-means clustering

### Dimensionality of Datasets

- Univariate (One variable, histogram or boxplots for example)
- Bivariate (two variables, scatterplot for example)
- Multivariate (two or more, bubble chart for example)

### EDA-Outliers

- How can we detect outliers
- One way is with the boxplot. [End of whiskers is calculates as Q1-1.5*IQR or Q3+1.5 IQR]
- another way is with the normal distribution. Standarizing Z-score [with thi one we consider 98% more or less of the data]

### Rescaling Data

- Normalization of the data (Min-Max scaling), rescales the data between 0 and 1
- Standardizing: mean of zero and standard deviation of 1

### Simple and Multiple linear regression

- OLS: Ordinary Least Squared Method.
- there is always an error

### Multi-Collinearity Issue

- Make sure that the regression is stable enough that you cannot rely on it. It confuses the betas of the regression.
- If it is perfectly correlated your model is unstable, betas could interchange if you run it again.
- Drop correlated variables. What is highly correlated.
- Correlation matrix is just to raise a flag.
- you can use the Variance Inflation Factor (VIF), it should be less than 5 or 10. It tries to predict one variable using another one. This is for all the features

### F Statistics

- Fstats says what is the odd that all of this variables are equal to 0, if F is large it tells you that at least one of them is greater than 0.
